```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

#read the CSVs in, sorted
autoSales = pd.read_csv('TOTALNSA.csv',usecols=['DATE','TOTALNSA'], parse_dates=['DATE'])

autoSales.head()

#set date as index
autoSales.set_index('DATE',inplace=True)


fig, ax = plt.subplots(figsize=(15,7))

ax.set_title('Auto Sales over Time')
ax.set_ylabel('Auto Sales')
ax.set_xlabel('Year')
ax.xaxis.set_major_locator(mdates.YearLocator()) #bin by year
ax.xaxis.set_major_formatter(mdates.DateFormatter('%y')) #format date

ax.plot(autoSales.index, autoSales['TOTALNSA'],color='#ef4d4d', alpha=0.9)

plt.show()
```
